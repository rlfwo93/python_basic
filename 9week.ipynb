```python
# 33

while True:
  
  sel = input("+,-,*,%,/  중에 선택해주세요 ~  : " )
  first,second = map(int,input("값을 입력 : ").split(","))
  answer =int(input("예상되는 정답을 입력해주세요 ! :"))
  sum = 0

  if sel =="+":
    sum = first + second

  elif sel == "-":
    sum = first - second
         
  elif sel =="*":
    sum = first * second
    
  elif sel == "/":
    sum = first / second
        
  if sum == answer:
    print("참 잘했네요 ~")
  else:
    print("공부 더 하세요 ~")
    
  br=input("계속 하시겠습니까 ? 네 or 아니오")
  if br == "네":
    continue
  else:
    break
      
```
